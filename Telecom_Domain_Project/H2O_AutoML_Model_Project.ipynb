```python
import h2o
from h2o.automl import H2OAutoML
h2o.init(max_mem_size='16G')
```

```python
# Loading the dataset
dataset = h2o.import_file("/content/Telecom.xlsx - in.csv")
```

```python
dataset.head()
```

```python
## splitting the data into training and test

df_train, df_test = dataset.split_frame(ratios=[.8])
```

```python
df_train
```

```python
df_test
```

```python
y = 'Churn' # Dependent variable
x = dataset.columns
x.remove(y)
```

```python
x.remove('customerID')
```

```python
x
```

```python
y
```

1.max_runtime_secs: This argument specifies the maximum time that the AutoML process will run for.

2.max_models: Specify the maximum number of models to build in an AutoML run, excluding the Stacked Ensemble models. Defaults to NULL/None

3.seed: Integer. Set a seed for reproducibility

```python
# Building H2O AutoML model
aml = H2OAutoML(max_runtime_secs=300, max_models=10, seed=10, verbosity='info', nfolds=2)
```

```python
## Trainin data

aml.train(x=x, y=y, training_frame=df_train)
```

```python
lb = aml.leaderboard
lb
```

```python
model_ids = list(aml.leaderboard['model_id'].as_data_frame().iloc[:,0])
```

```python
model_ids
```

```python
### test data

aml.leader.model_performance(df_test)
```

```python
# Hyper parameter tuning
h2o.get_model([mid for mid in model_ids if "StackedEnsemble" in mid][0])
```
